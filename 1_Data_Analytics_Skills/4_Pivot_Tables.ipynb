```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

print(pd.__version__)

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')

# Convert dataset to pandas DataFrame
df = dataset['train'].to_pandas()

# Change string to datetime ('job_posted_date' column)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(index='job_title_short', aggfunc='size')
#creates one-way pivot table with counts of job titles
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
#create a one-way pivot table finding median yearly salary
```

```python
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
#creates a 2 way pivot table for median salary per job for each country
#change orientation of 2 way pivot table by switching index and columns
```

```python
top_countries = df['job_country'].value_counts().head(6).index
top_countries
```

```python
df_job_country_salary = df.pivot_table(
    values='salary_year_avg', 
    index='job_country', 
    columns='job_title_short', 
    aggfunc='median'
    )
```

```python
df_job_country_salary.head(5)
```

```python
df_job_country_salary_top = df_job_country_salary.loc[top_countries]

```

```python
df_job_country_salary_top
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_job_country_salary_top_3_jobs = df_job_country_salary_top[job_titles]
df_job_country_salary_top_3_jobs
```

```python
#plot pivot table
df_job_country_salary_top_3_jobs.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Median Salary ($USD)')
plt.title('Median Salary for 3 Data Jobs by Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```
