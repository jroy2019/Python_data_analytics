```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

print(pd.__version__)

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')

# Convert dataset to pandas DataFrame
df = dataset['train'].to_pandas()

# Change string to datetime ('job_posted_date' column)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()
```

```python
df_usa.head()
```

```python
df_usa.reset_index(inplace=True, drop=True)
```

```python
df_usa.head()
```

```python
df_usa['job_month'] = df_usa['job_posted_date'].dt.strftime('%B')
```

```python
df_usa['job_month'].head()
```

```python
df_usa.head()
```

```python
df_pivot_usa = df_usa.pivot_table(index='job_month',
                                  columns='job_title_short',
                                  aggfunc='size')
```

```python
df_pivot_usa
#unordered (months not chronological)
```

```python
df_pivot_usa.reset_index(inplace=True)
```

```python
df_pivot_usa['month_num'] = pd.to_datetime(df_pivot_usa['job_month'], format='%B').dt.month
```

```python
df_pivot_usa['month_num'].head()
```

```python
df_pivot_usa.head()
```

```python
df_pivot_usa.sort_values(by='month_num', ascending=True, inplace=True)
```

```python
df_pivot_usa
```

```python
df_pivot_usa.set_index('job_month', inplace=True)
```

```python
df_pivot_usa.drop(columns=['month_num'], inplace=True)
```

```python
df_pivot_usa
```

```python
tech_jobs_usa_pivot = pd.read_csv('https://lukeb.co/software_csv')
```

```python
tech_jobs_usa_pivot.set_index('job_posted_month', inplace=True)
```

```python
tech_jobs_usa_pivot
```

Merge()
combine 2 dfs on a common columns

```python
data_jobs_usa_pivot = df_pivot_usa.copy()
df_usa_jobs_pivot = data_jobs_usa_pivot.merge(right=tech_jobs_usa_pivot, left_index=True, right_index=True)
df_usa_jobs_pivot
```

```python
top_5_jobs = (df_usa_jobs_pivot
              .sum()
              .sort_values(ascending=False)
              .head(5)
              .index
              .to_list())

top_5_jobs
```

```python
df_usa_jobs_pivot[top_5_jobs].plot(kind='line')
plt.title('Job Posting of Popular Data and Tech Jobs in USA')
plt.xlabel('2023')
plt.ylabel('Number of Job Postings')
plt.legend(loc='upper right')
plt.ylim(0,22000)
plt.show()
```
