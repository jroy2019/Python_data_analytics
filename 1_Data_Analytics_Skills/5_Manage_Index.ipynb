```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

print(pd.__version__)

# Load Data
dataset = load_dataset('lukebarousse/data_jobs')

# Convert dataset to pandas DataFrame
df = dataset['train'].to_pandas()

# Change string to datetime ('job_posted_date' column)
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(15)
```

```python
df.index
```

```python
df.index.name = 'job_index'
```

```python
df.sample(15)
#index row has name
```

```python
df.index.dtype
```

```python
df_usa = df[df['job_country'] == 'United States']
```

reset_index() <br>
reset index of a df from 1 to last index

```python
df_usa.head()
df_usa.reset_index(inplace=True)
```

```python
df_usa.head()
```

set_index() <br>
Changes index to index of another column in df

```python
df_usa.set_index('job_index', inplace=True)
```

```python
df_usa.head()
```

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median','min','max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False, inplace=True)
median_pivot
```

sort_index() <br>
sorts an index no column name like sort_values

```python
median_pivot.sort_index(inplace=True)
median_pivot #alpha order of first column by default
#sort index is good to sort by index (no column name needed)
```
